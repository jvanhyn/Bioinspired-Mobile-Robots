```python
# (a) Helper function cross_matrix(v)

def cross_matrix(v):
    vx = v[0].item()
    vy = v[1].item()
    vz = v[2].item()

    return np.array([[0,-vz,vy],
                     [vz,0,-vx,],
                     [-vy,vx,0]]);
```

```python
# (b) Helper function uncross_matrix(V)

def uncross_matrix(V):
    vx = V[2,1].item()
    vy = V[0,2].item()
    vz = V[1,0].item()
    
    return np.array([[vx],
                     [vy],
                     [vz]]);
```

```python
# (c) Dynamics function omega_dot(t, w, I_B)

def w_dot(t, w, I_B):
    return -(np.linalg.inv(I_B) @ (cross_matrix(w) @ I_B @ w.transpose())).transpose()
```

```python
# Explicit Euler Integration function
def ee(f,I,x0,dt,N):
    x = [x0]
    t = [0]
    for i in range(N-1):
        x.append(x[i] + f(x[i],I) * dt)
        t.append(i*dt)
    return x, t

# Running the integration
h = 0.01 # ms
N = 1000  #
w1, t = ee(w_dot,I_B,w1_0,h,N)
w2, t = ee(w_dot,I_B,w2_0,h,N)
```
